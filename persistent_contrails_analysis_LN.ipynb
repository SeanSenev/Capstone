# modelling

```python
#define X & y 
X = modeling_df2.drop(columns=["Persistent contrail length (km)", "Contrail age, Mean (h)"])
y1 = modeling_df2["Persistent contrail length (km)"]
y2 = modeling_df2["Contrail age, Mean (h)"]
```

```python
#split into train vs val as 8:2
from sklearn.model_selection import train_test_split

X_train, X_val, y1_train, y1_val, y2_train, y2_val = train_test_split(X, y1, y2, test_size=0.2, random_state=42)
X_train.shape, X_val.shape
```

```python
#feature selections using randomforest - takes 30mins!!!
from sklearn.ensemble import RandomForestRegressor
rf_selector = RandomForestRegressor()
rf_selector.fit(X_train, y1_train)   #using y1 as target

important_features = rf_selector.feature_importances_
top_10_indices = important_features.argsort()[-10:][::-1]
X_train_selected = X_train.iloc[:, top_10_indices]
X_val_selected = X_val.iloc[:, top_10_indices]
```

```python
#standardisation
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_selected)
X_val_scaled = scaler.transform(X_val_selected)
```

```python
#y combined 

from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

def train_and_evaluate(model, X_train, y_train, X_val, y_val):
    model.fit(X_train, y_train)
    predictions = model.predict(X_val)
    
    mse = mean_squared_error(y_val, predictions)
    rmse = mse**0.5
    
    return rmse

#linear regression
lr_rmse = train_and_evaluate(LinearRegression(), X_train, y_train, X_val, y_val)
print(f"Linear Regression RMSE: {lr_rmse}")

#random forest
rf_rmse = train_and_evaluate(RandomForestRegressor(), X_train, y_train, X_val, y_val)
print(f"Random Forest RMSE: {rf_rmse}")
```
