```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print(data.describe())
```

```python
print(data.info())
```

```python
print(data.isnull().sum())
```

```python
numerical_features = data.select_dtypes(include=['float64', 'int64'])
numerical_features.hist(bins=50, figsize=(20,15))
plt.tight_layout()
plt.show()
```

```python
# Boxplots (showing for only a few columns for the sake of brevity)
cols_to_show_in_boxplot = ['Total flight dist (km)', 'Total fuel burn (kg)', 'Aircraft mass mean (kg)']
for column in cols_to_show_in_boxplot:
    sns.boxplot(data[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
# Correlation Matrix
correlation = numerical_features.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

# Barplot for 'Direction' column
sns.countplot(data['Direction'])
plt.title("Count of Flights for each Direction")
plt.show()
```
