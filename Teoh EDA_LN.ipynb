```python
data['First known waypoint time (UTC)'] = pd.to_datetime(data['First known waypoint time (UTC)'])
data['Last known waypoint time (UTC)'] = pd.to_datetime(data['Last known waypoint time (UTC)'])
data['Duration'] = data['Last known waypoint time (UTC)'] - data['First known waypoint time (UTC)']
data['Duration'].describe()  #max duration 09:23:10 <-> min duration 00:13:26
```

```python
aircraft_type = data.groupby('ICAO Aircraft type designator')
aircraft_type.size().sort_values(ascending=False)
```

```python
#ave duration for each aircraft type
def calculate_duration(df):
    durations = pd.to_datetime(df['Last known waypoint time (UTC)']) - pd.to_datetime(df['First known waypoint time (UTC)'])
    return durations.mean()
duration = aircraft_type.apply(calculate_duration)

#flight count/initial contrail count/persistent contrail count for each aircraft type
flight_count = aircraft_type.size()
initial_contrail_count = aircraft_type.apply(lambda df: (df['Initial contrail length (km)'] > 0).sum())
persistent_contrail_count = aircraft_type.apply(lambda df: (df['Persistent contrail length (km)'] > 0).sum())

#concat to summary table
summary_df = pd.DataFrame({
    'ave_duration': duration,
    'ave_flight_count': flight_count,
    'ave_initial_contrail_count': initial_contrail_count,
    'aver_persistent_contrail_count': persistent_contrail_count
})
```

```python
# summary_df
summary_df.sort_values(by='ave_duration', ascending=False)  #sort by individual cols
```

```python
cols_ave = [
    'Total flight dist (km)', 'Total fuel burn (kg)', 'Mean overall propulsion efficiency', 
    'Aircraft mass mean (kg)', 'Mean nvPM number EI (kg-1)', 'Initial contrail length (km)',
    'RHi initial, Mean', 'RHi lifetime, Mean', 'Temp initial, Mean (K)', 'Temp SAC, Mean (K)',
    'Persistent contrail length (km)', 'Contrail altitude initial, Mean (ft)', 
    'Contrail altitude lifetime, Mean (ft)', 'Contrail age, Mean (h)', 
    'Ice number initial, Mean (m-1)', 'Ice number lifetime, Mean (m-1)', 
    'Ice vol mean radius, Mean (um)', 'Tau contrail, Mean', 'Tau cirrus, Mean', 
    'Total contrail EF (J)'
]

averages_df = aircraft_type[cols_ave].mean()
# summary_df = summary_df.join(averages_df)

averages_df.columns = ['ave_' + col for col in averages_df.columns]
averages_df  #ave values by aircraft type
```
