# split data & feature selection

```python
#define X & y 

modeling_df2 = modeling_df2.sample(frac=1, random_state=42).reset_index(drop=True)  #shuffle data

X = modeling_df2.drop(columns=["Persistent contrail length (km)", "Contrail age, Mean (h)"])
y1 = modeling_df2["Persistent contrail length (km)"]
y2 = modeling_df2["Contrail age, Mean (h)"]
```

```python
#split into train vs val vs test as 8:1:1

from sklearn.model_selection import train_test_split
#train:test as 8:2
X_train, X_temp, y1_train, y1_temp, y2_train, y2_temp = train_test_split(
    X, y1, y2, test_size=0.2, random_state=42)

#val:test as 1:1
X_val, X_test, y1_val, y1_test, y2_val, y2_test = train_test_split(
    X_temp, y1_temp, y2_temp, test_size=0.5, random_state=42)  # random_state can be different or same

X_train.shape, X_val.shape, X_test.shape
```

```python
#feature selections using randomforest - takes 30mins!!!
from sklearn.ensemble import RandomForestRegressor
rf_selector = RandomForestRegressor()  #takes 30mins!!!
rf_selector.fit(X_train, y2_train)   #using y2 as target ie contrail age

important_features = rf_selector.feature_importances_
top_10_indices = important_features.argsort()[-10:][::-1]

#X input for RF
X_train_selected = X_train.iloc[:, top_10_indices]
X_val_selected = X_val.iloc[:, top_10_indices]
X_test_selected = X_test.iloc[:, top_10_indices]
```

```python
#top 10 features based on RandomForestRegressor
top_features = X_train.columns[top_10_indices]
top_features
```
